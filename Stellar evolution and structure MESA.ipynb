```python
mu_e = 1.2
mu = np.average(m14.hist.data['center_mu'])
rho = np.logspace(-10,10,10000)

def Trad(rho, mu):
    return 3.2*10**7 * mu**(-1/3) * rho**(1/3)

def TNR(rho, mu, mu_e):
    return 1.21*10**5 * mu * mu_e**(-5/3) * rho**(2/3)

def TER(rho, mu, mu_e):
    return 1.50*10**7 * mu * mu_e**(-4/3) * rho**(1/3)

def crazy(rho, mu_e):
    return (9.7*10**5 * mu_e) + rho*0
```

```python
m14 = mp.MESA()
m14.loadHistory(f='C:/Users/DavidKonijn/MESA-Docker/docker_work/work/logsfolder/LOGS_14')

fig, ax1 = plt.subplots(figsize=(10,8))
ax1.plot(m14.hist.data['log_cntr_Rho'][:220], m14.hist.data['log_cntr_T'][:220], color= 'black')
ax1.plot(np.log10(rho), np.log10(Trad(rho, mu)), linestyle='dashed', color = 'black', alpha=0.5)
ax1.plot(np.log10(rho)[:8180], np.log10(TNR(rho, mu, mu_e))[:8180], linestyle='dashed', color = 'black', alpha=0.5)
ax1.plot(np.log10(rho)[8172:], np.log10(TER(rho, mu, mu_e))[8172:], linestyle='dashed', color = 'black', alpha=0.5)
ax1.plot(np.log10(crazy(rho, mu_e))[:9580], np.log10(rho)[:9580], linestyle='dashed', color = 'black', alpha=0.5)

# plt.axvline(x=np.log10(9.7*10**5 * mu_e), linestyle='dashed', color = 'black')
# ax1.set_title('Temperature Density plane', fontsize=20)
ax1.set_xlabel('Log('r'$\rho$) in g/$cm^3$)', fontsize = 20)
ax1.set_ylabel('Log(T) in Kelvin', fontsize = 20)

ax1.fill_between(np.log10(rho), 100,  np.log10(Trad(rho, mu)), color='#F26419', alpha=0.2)
ax1.fill_between(np.log10(rho)[:8180], np.log10(Trad(rho, mu))[:8180], np.log10(TNR(rho, mu, mu_e))[:8180], color='#FFEE2D', alpha=0.2)
ax1.fill_between(np.log10(rho)[:8027], -1, np.log10(TNR(rho, mu, mu_e))[:8027], color='#33658A', alpha=0.2)
ax1.fill_between(np.log10(rho)[8027:8181], -1, np.log10(TNR(rho, mu, mu_e))[8027:8181], color='white', alpha=0.1)
ax1.fill_between(np.log10(rho)[8187:8800], 1, np.log10(TER(rho, mu, mu_e))[8187:8800], color='white', alpha=0.1)
ax1.fill_between(np.log10(rho)[8183:], np.log10(TER(rho, mu, mu_e))[8183:], np.log10(Trad(rho, mu))[8183:], color='#FFEE2D', alpha=0.2)


plt.text(0.7, 9, 'Radiation', fontsize = 15, fontfamily='sans-serif')
plt.text(3.0, 7.8, 'Ideal Gas', fontsize = 15, fontfamily='sans-serif')
plt.text(5.3, 7.9, 'Degenerate', fontsize = 15, fontfamily='sans-serif')
plt.text(5, 7.7, 'NR', fontsize = 15, fontfamily='sans-serif')
plt.text(6.4, 7.7, 'ER', fontsize = 15, fontfamily='sans-serif')


ax1.annotate("Central Helium", xy=(3.3, 8.2), xytext=(3.8, 8.3),arrowprops=dict(arrowstyle="->"))
ax1.annotate("", xy=(4.72, 8.6), xytext=(4.3, 8.37),arrowprops=dict(arrowstyle="->"))
ax1.annotate("Central Hydrogen", xy=(1.4, 7.6), xytext=(1.8, 7.52),arrowprops=dict(arrowstyle="->"))
ax1.annotate("", xy=(3.2, 8.05), xytext=(2.3, 7.6),arrowprops=dict(arrowstyle="->"))

plt.xlim(0, 7)
plt.ylim(7.5, 10)

ax1.fill_between([0,1],0,[1,0], color='#FFEE2D', label='Ideal Gas', alpha = 0.5)
ax1.fill_between([0,1],0,[1,0], color='#F26419', label='Radiation')
ax1.fill_between([0,1],0,[1,0], color='white', label='Extreme Relativistic')
ax1.fill_between([0,1],0,[1,0], color='#33658A', label='Non-Relativistic')

leg = plt.legend()
leg.get_frame().set_edgecolor('black')      
leg.get_frame().set_facecolor('none')
plt.savefig('Temperature_Density_plot.pdf')
plt.show()
```

```python
import numpy as np
import mesaPlot as mp
import matplotlib.pyplot as plt
import mkipp
```

```python
fig = plt.figure(figsize=(15,10))
axis = plt.gca()

kipp_plot = mkipp.kipp_plot(mkipp.Kipp_Args(logs_dirs = ['LOGS_14'], contour_colormap = plt.get_cmap("YlOrRd")), axis = axis)
axis.set_xlabel("Model Number", fontsize = 20)
axis.set_ylabel("Mass (solar masses)", fontsize = 20)
axis.set_xlim(0,1100)

axis.annotate("Central hydrogen", xy=(50, 13.2), xytext=(50, 13.3))
axis.annotate("burning ends", xy=(50, 12.9), xytext=(50, 12.9))

axis.annotate("Central helium", xy=(310, 13.2), xytext=(310, 13.3))
axis.annotate("burning starts", xy=(310, 12.9), xytext=(310, 12.9))

axis.annotate("Convective envelope", xy=(530, 10.2), xytext=(530, 10.2))
axis.annotate("develops", xy=(530, 9.9), xytext=(530, 9.9))

axis.annotate("Central helium", xy=(750, 4.2), xytext=(750, 4.2))
axis.annotate("burning ends", xy=(750, 3.9), xytext=(750, 3.9))

plt.axvline(x=200, c='green', linestyle='dashed', linewidth = 2.5)
plt.axvline(x=300, c='green', linestyle='dashed', linewidth = 2.5)
plt.axvline(x=520, c='green', linestyle='dashed', linewidth = 2.5)
plt.axvline(x=880, c='green', linestyle='dashed', linewidth = 2.5)

plt.savefig('Kippenhahn14.pdf')

print(kipp_plot.xlims)
```
